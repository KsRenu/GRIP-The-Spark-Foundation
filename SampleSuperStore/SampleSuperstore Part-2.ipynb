# Analysing SampleSuperstore dataset and deriving the necessary insights-                                                           Part-2

## Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path="SampleSuperstore.csv"
df=pd.read_csv(file_path)
```

## Double Column Analysis

##### As overall place details can be derived from city and state itself ,postal code isn't necessary.

```python
df.drop('Postal Code',axis=1,inplace=True)
```

## Ship Mode and Segment

```python
sns.catplot(x="Segment", hue="Ship Mode", kind="count",data=df)

```

## Ship Mode and Region

```python
sns.countplot(hue="Ship Mode", x="Region",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Ship Mode and Category

```python
sns.countplot(hue="Ship Mode", x="Category",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Ship Mode and Sub-Category

```python
sns.countplot(hue="Ship Mode",y="Sub-Category",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Ship Mode and Quantity

```python
sns.countplot(hue="Ship Mode", x="Quantity",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

# Segment and Region

```python
sns.countplot(hue="Segment", x="Region",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Segment and Category

```python
sns.countplot(hue="Segment", x="Category",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Segment and Sub-Category

```python
sns.countplot(hue="Segment", y="Sub-Category",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Segment and Quantity

```python
sns.countplot(hue="Segment", x="Quantity",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```

## Region and Category

```python
sns.countplot(hue="Category", x="Region",data=df,palette="Set3")
sns.set_theme(style="darkgrid")
```
